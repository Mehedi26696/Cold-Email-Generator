<b>API Key Set Up

```python
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Access the API key
GROQ_API_KEY = os.getenv("GROQ_API_KEY")

```

<b>Conncetion Testing

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0,
    groq_api_key = GROQ_API_KEY,
    max_tokens=None,
    timeout=None,
    max_retries=2,
)
```

```python
response = llm.invoke("Write 10 Football Players name")
print(response.content)
```

<b>Scrapping Job Posts

```python
from langchain_community.document_loaders import WebBaseLoader

 

loader = WebBaseLoader("https://jobs.bdjobs.com/jobsearch.asp?fcatId=8&icatId=")
```

```python
page_data = loader.load().pop().page_content
```

```python
print(page_data)
```

<b>Extract Necessary Information

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format **under a single top-level key called `jobs`**.
    Each job must include: `role`, `experience`, `skills`, and `description`.
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):    
    """
)
```

```python
chain_extract = prompt_extract | llm 
res = chain_extract.invoke(input={'page_data':page_data})
type(res.content)
```

```python
print(res.content)
```

<b>Convert in Json Format

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
type(json_res)
```

<b>Store Portfolio Info At Chromadb

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```

<b>Cold Email Generate

```python
prompt_email = PromptTemplate.from_template(
        """
        ### JOB DESCRIPTION:
        {job_description}
        
        ### INSTRUCTION:
        You are Mehedi, a business development executive at Therap. Therap is an AI & Software Consulting company dedicated to facilitating
        the seamless integration of business processes through automated tools. 
        Over our experience, we have empowered numerous enterprises with tailored solutions, fostering scalability, 
        process optimization, cost reduction, and heightened overall efficiency. 
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of Therap 
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Therap's portfolio: {link_list}
        Remember you are Mehedi, BDE at Therap. 
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):
        
        """
        )

```

```python
chain_email = prompt_email | llm
res = chain_email.invoke({"job_description": str(job1), "link_list": links})
print(res.content)
```
